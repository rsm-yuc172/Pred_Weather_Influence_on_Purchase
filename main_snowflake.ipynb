Tree Variable
The tree variable is an instance of the ElementTree class. It represents the entire XML document and provides methods to access the root element and other global document properties. You cannot directly print it to see the XML contents, but you can write it back to an XML string or file to examine its structure.

Root Variable
The root variable is an instance of the Element class. It represents the root element of the XML document. An Element object has several useful properties and methods:


tag: the name of the element (e.g., <element>'s tag is "element").
text: the text inside the element.
attrib: a dictionary containing the element's attributes.
tail: the text that follows the element, up to the next element in the XML tree.

```python
import xml.etree.ElementTree as ET

cs.execute("""
CREATE OR REPLACE TABLE PJ_DB.PJ_SCHEMA.Supplier_Transactions(
    SupplierTransactionID INTEGER,
    SupplierID INTEGER,
    TransactionTypeID INTEGER,
    PurchaseOrderID INTEGER,
    PaymentMethodID INTEGER,
    SupplierInvoiceNumber INTEGER,
    TransactionDate DATE,
    AmountExcludingTax FLOAT,
    TaxAmount FLOAT,
    TransactionAmount FLOAT,
    OutstandingBalance FLOAT,
    FinalizationDate DATE,
    IsFinalized INTEGER,
    LastEditedBy INTEGER,
    LastEditedWhen TIMESTAMP)
    """)

# Parse XML 
tree = ET.parse('/home/jovyan/sql_data/Project/Case Data/Supplier_Transactions_XML.xml')
root = tree.getroot()

# Loop through each row in the XML file
for row in root:
    values = []
    for elem in row:
        # Append each value to the list, escaping and quoting string values
        if elem.text is not None:
            values.append("'" + elem.text.replace("'", "''") + "'" if elem.tag in ["TransactionDate", "FinalizationDate", "LastEditedWhen"] else elem.text)
        else:
            values.append("NULL")
# Create an SQL insert statement
    sql = f"INSERT INTO PJ_DB.PJ_SCHEMA.Supplier_Transactions VALUES ({','.join(values)})"

    # Execute the SQL insert statement
    cs.execute(sql)
```

```python
# pushing data from stage to table zipcodes
cs.execute("""
COPY INTO zipcodes 
FROM @zipcode_stage 
FILE_FORMAT = CSV_FORMAT ON_ERROR = 'CONTINUE'
""")
```

```python
# condensing weather data to get single row for each day and station id into a table called noaa_gsod
cs.execute("""
create or replace table noaa_gsod
cluster by (station_id, date)
as (
    select *
    from (
        select "Stations", "Date", "Stations Name", "Country", "Stations Latitude", "Stations Longitude", "Indicator Name", "Value"
        from ENVIRONMENT_DATA_ATLAS.ENVIRONMENT.NOAACD2019R
        where "Measure"='M1'
              and "Country"='US'
    )
    pivot(max("Value") for "Indicator Name" in ('Mean visibility (miles)','Maximum temperature (Fahrenheit)','Mean dew point (Fahrenheit)','Maximum wind gust (Number)','Minimum temperature (Fahrenheit)','Maximum sustained wind speed (knots)','Mean wind speed (knots)','Mean station pressure (millibars)','Precipitation amount (inches)','Mean temperature (Fahrenheit)','Mean sea level pressure (millibars)','Snow depth (inches)'))
    as p(station_id, date, name, country_fips, lat, long, visibility, max, dew, wind_max, min, wind_sustained_max, wind_mean, pressure, rain, temp, pressure_sea, snow_depth)
)
""")
```

```python
# merging noaa_gsod and zipcodes and retaining shortest distance between latitude and longitude for both
# output is matching zip code to weather stations
cs.execute("""
create or replace table stations_zipcode
as (
    select zip_code, a.station_id, st_distance(st_makepoint(a.long, a.lat), st_makepoint(b.longitude, b.latitude)) distance
    from noaa_gsod as a
    join zipcodes as b
    on b.country=a.country_fips
    and st_distance(st_makepoint(a.long, a.lat), st_makepoint(b.longitude, b.latitude)) < 50000
    qualify row_number() over(partition by a.station_id order by distance) = 1
)
""")
```

```python
# join the previous table with noaa_gsod for temperature and date - part a done
cs.execute("""
CREATE OR REPLACE TABLE zip_temp_all AS
SELECT b.date, b.temp, a.zip_code
FROM stations_zipcode as a
JOIN noaa_gsod as b
ON a.station_id = b.station_id
""")
```

```python
#part b - get unique date, zip, and temp for zipcodes in supplier_case

cs.execute("""
CREATE OR REPLACE TABLE supplier_zip_code_weather AS
SELECT DISTINCT a.postalpostalcode AS zip, b.date, b.temp
FROM supplier_case as a
LEFT JOIN zip_temp_all as b
ON a.postalpostalcode = b.zip_code
""")
```
